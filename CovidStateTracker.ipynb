```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from pandas.io.json import json_normalize
URL='https://covidtracking.com/api/states/daily?state='
STATES=["IL", "KY", "MO", "TN"]
STATEURLS=[]
REQUESTSGET=[]
for i in STATES:
    STATEURLS.append(i + "=" +URL+i)
for i in STATEURLS:
    i
    STATE=i[-2:]
    STATEGET = requests.get(i[+3:])
    STATE = json_normalize(STATEGET.json())
    DATE=STATE['date']
    POSITIVE=STATE['positive']
    plt.plot(DATE, POSITIVE, label="Positive " + i[-2:])
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.5)
plt.grid()
plt.show()

for i in STATEURLS:
    i
    STATE=i[-2:]
    STATEGET = requests.get(i[+3:])
    STATE = json_normalize(STATEGET.json())
    DATE=STATE['date']
    DEATH=STATE['death']
    plt.plot(DATE, DEATH, label="Deaths " + i[-2:])
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.5)
plt.grid()
plt.show()
```
